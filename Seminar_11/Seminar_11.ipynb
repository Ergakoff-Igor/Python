```python
from sympy import *
from sympy.plotting import plot

x = symbols('x')
p1 = plot(x*x, show = false)
p2 = plot(5*x, show = false)
p1.append(p2[0])
p1.show()
```

```python
from sympy import *
from sympy.plotting import plot

x = symbols('x')
p1 = plot(-12*x*x*x*x*sin(cos(x)) - 18*x*x*x+5*x*x + 10*x - 30, show = false)
p2 = plot(5*x, show = false)
p1.append(p2[0])
p1.show()
```

```python
x = symbols('x')
p1 = plot(x*x, show = False)
p2 = plot(32*(-x)**0.5, show = False)
p3 = plot(-32*x**0.5, show = False)
p4 = plot(-32*(-x)**0.5, show = False)
p5 = plot(32*x**0.5, show = False)
p6 = plot(-(x*x), show = False)
p1.append(p2[0])
p1.append(p3[0])
p1.append(p4[0])
p1.append(p5[0])
p1.append(p6[0])
p1.show()
```
