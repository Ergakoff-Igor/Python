дана функция: __f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30__

1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0

```python
from sympy import *
from sympy.plotting import plot
```

Вывод функции в математическом виде:

```python
x = Symbol('x')
y = -12 * (x**4) * sin(cos(x)) - 18 * (x**3) + 5 * (x**2) + 10*x - 30
y
```

1. Опредеение корней:

```python
nonlinsolve([y], x)
```

2. Интервалы, на которых функция возрастает:

```python
solveset(y>0)
```

3. Интервалы, на которых функция убывает:

```python
solveset(y<0)
```

4. График на интервале -10000, 10000:

```python
graph = plot(y, (x,-10000, 10000))
```

6. вершины:

```python
pr_expr = y.diff(x)
nonlinsolve([pr_expr],[x])
```

7. промежутки, на которых f > 0:

```python
solveset(y>0)
```

8. промежутки, на котором f < 0:

```python
solveset(y<0)
```
